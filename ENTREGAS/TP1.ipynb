### ***Procedimiento***
Para empezar importamos las librerias a utilizar

```python
import numpy as np
import matplotlib.pyplot as plt
from math import pi
from scipy import signal
```

Realizamos la primera señal para ello definimos los parametros a utilizar, luego realizamos el vector tiempo y utilizando la funcion *"square"* de la libreria scipy realizamos la señal cuadrada a analizar. Por ultimo, graficamos dicha señal.

```python
# parametros de la señal
frec = 0.5 #[Hz]
amplitud = 0.5 #[V]
duracion = 4 #[s]
dutyCycle = 0.10 
nivelDC = 1.4 #[V]
fs = 1000 #[Hz]

vectorTemp = np.linspace(-0.005, duracion, int(duracion * fs), endpoint= False) # vector de tiempo

# señal cuadrada
senial = amplitud * signal.square(2 * pi * frec * vectorTemp, dutyCycle)
senial = senial + nivelDC # ponemos un nivel de continua.

# grafica
fig = plt.gcf()
fig.set_size_inches(18.5,10.5)

plt.plot(vectorTemp, senial, color='orange', lw = 5)
plt.title('Señal Cuadrada 1')
plt.xlabel('Tiempo [s]')
plt.ylabel('Amplitud [V]')
plt.grid()
plt.show()
```

De igual forma que en la señal anterior creamos al otra señal cuadrada

```python
# parametros de la señal
frec_2 = 0.5 #[Hz]
duracion_2 = 4 #[s]
dutyCycle_2 = 0.10 
nivelDC_2 = 0.6 #[V]

vectorTemp_2 = np.linspace(-0.005, duracion_2, int(duracion_2 * fs), endpoint= False) # vector de tiempo

# señal cuadrada
senial_2 = -amplitud * signal.square(2 * pi * frec_2 * vectorTemp_2, dutyCycle_2)
senial_2 = senial_2 + nivelDC_2 # ponemos un nivel de continua.

# grafica
fig = plt.gcf()
fig.set_size_inches(18.5,10.5)

plt.plot(vectorTemp_2, senial_2, color='red',lw = 5)
plt.title('Señal Cuadrada 2')
plt.xlabel('Tiempo [s]')
plt.ylabel('Amplitud [V]')
plt.grid()
plt.show()
```

Finalmente calculamos el valor medio, eficaz y eficaz de alterna para ambas señales 

```python
# Valor medio de la señal cuadrada 1 
valorMedio_1 = np.mean(senial)
print('Valor medio de la señal 1: ', valorMedio_1)

# Valor medio de la señal cuadrada 2 
valorMedio_2 = np.mean(senial_2)
print('Valor medio de la señal 2: ', valorMedio_2)

# Valor eficaz de la señal cuadrada 1 
valorRMS_1 = np.sqrt(np.mean(senial**2))
print('Valor eficaz de la señal 1: ', valorRMS_1)

# Valor eficaz de la señal cuadrada 2 
valorRMS_2 = np.sqrt(np.mean(senial_2**2))
print('Valor eficaz de la señal 2: ', valorRMS_2)

# Valor eficaz de alterna de la señal cuadrada 1 
valorRMS_ac_1 = np.sqrt(np.mean(senial**2)-valorMedio_1**2)
print('Valor eficaz de alterna de la señal 1: ', valorRMS_ac_1)

# Valor eficaz de alterna de la señal cuadrada 2
valorRMS_ac_2 = np.sqrt(np.mean(senial_2**2)-valorMedio_2**2)
print('Valor eficaz de alterna de la señal 2: ', valorRMS_ac_2)
```

### ***Conclusión***
Se puede observar que aunque tiene una pequeña diferencia los valores obtenidos son practicamente iguales, esto es lo esperado puesto que aunque las señales se encuentran entre valores diferentes de tensión y la duración que tienen entre valores altos y bajos son inversas, en esencia las señales son similares puesto que sus periodos son los mismos, el *"duty cicle"* de ambas es el mismo y su amplitud es la misma (*1 [V]*) para ambas.
