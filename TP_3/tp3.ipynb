```python
def FFT(y,dt):
    # Realiza la FFT de la señal y
    # array(complex) FFT(array complex y, int n, float t)
    # Entradas:
    #          y: vector de longitud N (array D)
    #          dt: intervalo de muestreo de la señal y
    # Salida:
    #          Y: resultado de FFT de y (array D)
    #          frec: frecuencia asociadoas a FFT(y)
    
    n = len (y)
    Y = np.fft.fft(y) #normalizada
    frec = np.fft.fftfreq(n,dt)
    return Y, frec
```

```python
def cuadrada(senial):
    y = []
    for signal in senial:
        if signal == 1:
            valor_cuad = 1
        else:
            valor_cuad = 0
        y.append(valor_cuad)
    y = np.array(y)
    return(y)
```

```python

# Crear una señal de ejemplo
temp = np.linspace(0, 0.5, 500)
senal = np.sin(40 * 2 * np.pi * temp) + 0.5 * np.sin(90 * 2 * np.pi * temp)

# Calcular la FFT
fft = np.fft.fft(senal)

# Calcular las frecuencias para los valores de la FFT
N = len(senal)
frecuencias = np.fft.fftfreq(N)

# Graficar la señal original y la FFT
plt.figure()

plt.subplot(2, 1, 1)
plt.plot(temp, senal)
plt.title('Señal Original')

plt.subplot(2, 1, 2)
plt.plot(frecuencias, np.abs(fft))
plt.title('FFT de la Señal')

plt.tight_layout()
plt.show()
```
